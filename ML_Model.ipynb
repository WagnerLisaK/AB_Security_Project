# ML

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import hvplot.pandas
from sklearn.cluster import KMeans
```

```python
file = "MLInternetDataENCODED-LE.csv"
#file = "MLInternetDataENCODED-DUM.csv"
results_df = pd.read_csv(file)
results_df
```

```python
results_df.dtypes
```

```python
# Store value of K to plot
inertia = []
k = list(range(1, 11))
```

```python
# Looking for the best K
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(results_df)
    inertia.append(km.inertia_)
```

```python
# Plot elbow curve
elbow_data = {"k": k, "inertia": inertia}
elbow_df = pd.DataFrame(elbow_data)
elbow_df.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k)
```

```python
# Initialize model with K=3 as a starting point

model = KMeans(n_clusters=2, random_state=0)
model
```

```python
# Fitting model
model.fit(results_df)
```

```python
# Get predictions
predictions = model.predict(results_df)
print(predictions)
```

```python
# Create return DataFram with the predicted outcome

results_df['outcome'] = model.labels_
results_df
```

```python
outcome_counts = results_df.outcome.value_counts()
print(outcome_counts)
```

```python
# Plot for K=2
results_df.hvplot.scatter(x='UsrAgntSHORT', y = 'Time', by='outcome')
```
