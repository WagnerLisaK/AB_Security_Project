# ML - Encode

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
file = "MLInternetDataNOTencoded.csv"
results_df = pd.read_csv(file)
results_df = results_df.drop('Lat', axis = 1)
results_df = results_df.drop('Lng', axis = 1)
results_df
```

```python
results_df.dtypes
```

```python
target = results_df[results_df['IP_Address']=='70.81.197.137']
target
```

```python
target.to_csv('target.csv')
```

```python
encodedLE_df = results_df.copy()

# Encode within the same column
le=LabelEncoder()
encodedLE_df['IP_Address'] = le.fit_transform(encodedLE_df['IP_Address'])
encodedLE_df['DateTime'] = le.fit_transform(encodedLE_df['DateTime'])
encodedLE_df['Date'] = le.fit_transform(encodedLE_df['Date'])
encodedLE_df['Time'] = le.fit_transform(encodedLE_df['Time'])
encodedLE_df['Lat_Lng'] = le.fit_transform(encodedLE_df['Lat_Lng'])
encodedLE_df['gcl_wbr'] = le.fit_transform(encodedLE_df['gcl_wbr'])
encodedLE_df['bot_yn'] = le.fit_transform(encodedLE_df['bot_yn'])
encodedLE_df['Gbot_yn'] = le.fit_transform(encodedLE_df['Gbot_yn'])
encodedLE_df['UsrAgntSHORT'] = le.fit_transform(encodedLE_df['UsrAgntSHORT'])

encodedLE_df
```

```python
encodedLE_df.dtypes
```

```python
encodedLE_df.to_csv('MLInternetDataENCODED-LE.csv', index=False)
encodedLE_df.to_csv('EncodedLEIndexed.csv')
```

```python
# Encode with "get_dummies"

encodedDUM_df = results_df.copy()

encodedDUM_df = pd.get_dummies(results_df, columns=['IP_Address', 'DateTime', 'Date', 'Time', 'Lat_Lng', 'gcl_wbr', 'bot_yn', 'Gbot_yn', 'UsrAgntSHORT'])

encodedDUM_df
```

```python
encodedDUM_df.to_csv('MLInternetDataENCODED-DUM.csv', index=False)
```
